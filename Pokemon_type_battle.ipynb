```python
def opponent_chooser(pixel_number):
    corner_1 = [width, 1]
    corner_2 = [-1, width]
    corner_3 = [-1*width, 1]
    corner_4 = [-1*width, -1]
    
    edge_1 = [-1*width, 1, width]
    edge_2 = [-1*width, -1, width]
    edge_3 = [-1, 1, width]
    edge_4 = [-1*width, -1, 1]
    
    middle = [-1*width, -1, 1, width]
    
    if pixel_number == 0:
        opponent = pixel_number + random.choice(corner_1)
        
    elif pixel_number == width-1:
        opponent = pixel_number + random.choice(corner_2)
        
    elif pixel_number == (width*height)-width:
        opponent = pixel_number + random.choice(corner_3)
    
    elif pixel_number == (width*height)-1:
        opponent = pixel_number + random.choice(corner_4)
        
    elif pixel_number%height == 0:
        opponent = pixel_number + random.choice(edge_1)
    
    elif pixel_number%height == height-1:
        opponent = pixel_number + random.choice(edge_2)
        
    elif pixel_number < width:
        opponent = pixel_number + random.choice(edge_3)
        
    elif pixel_number > (width*height)-width:
        opponent = pixel_number + random.choice(edge_4)
        
    else:
        opponent = pixel_number + random.choice(middle)
    
    return opponent
```
